# Additional Dataframe Operations

```python
#"client_list.csv" ve "customer_id_list.csv" kullanın. Elde edilen veri çerçevesini şu şekilde atayın.mergeclients
df = pd.read_csv("client_list.csv")
df.columns = [x.upper() for x in df.columns]
ids = pd.read_csv("customer_id_list.csv")
clients = pd.merge(left=df, right=ids, how='right', on=['LAST_NAME','FIRST_NAME','GENDER','AGE'])
```

```python
df
```

```python
ids
```

```python
clients
```

```python
#Ausingand "purchase_log.csv" gerçekleştirin ve alt kümeyi yalnızca satın alma işlemi yapan istemcilerle sınırlayın. Elde edilen veri çerçevesini şu şekilde atayın.mergeclientsdetailed_sales
purchase = pd.read_csv("purchase_log.csv")
detailed_sales = pd.merge(left=clients, right=purchase, how='inner', on=['CUSTOMER_ID'])
```

```python
detailed_sales 
```

```python
#UseSatın alımlara en çok para harcayan müşteriyi bulmak için. Ne kadar harcadığını belirleyin. İPUCU: Satın alma işlemlerinde en çok para harcayan müşteriyi belirlemek için dilimleme uygulamadan önce ara veri çerçevesini usingas'tan kaydedin.groupbygroupbyspenders
spenders = detailed_sales.groupby(
    ['CUSTOMER_ID','FIRST_NAME','LAST_NAME','GENDER','AGE'], as_index=False)['PRICE'].sum()
spenders[spenders.PRICE==spenders.PRICE.max()]
```

```python
#(BONUS) En çok harcama yapan müşteri tarafından tam olarak hangi kalemlerin satın alındığını belirlemek için sorun 3'ün cevabını biraz değiştirin.
top_spender_id = spenders[spenders.PRICE==spenders.PRICE.max()].reset_index(drop=True).loc[0,'CUSTOMER_ID']
purchase.loc[purchase.CUSTOMER_ID==top_spender_id, 'ITEM_DESCRIPTION']
```

```python
#"df_out.csv" adlı bir csv dosyasını dizin olmadan kaydedin.detailed_sales
detailed_sales.to_csv("df_out.csv", index=False)
```
