```python
df=df.drop(["Navy Fleet Strengths value","Aircraft Carrier Fleet Strength value","Submarine Fleet Strength value","Destroyer Fleet Strength value","Frigate Fleet Strength value","Total Coastline Coverage value","Total Border Coverage value"],axis=1)
```

```python
from sklearn.cluster import KMeans # En yakın komşular
from sklearn.metrics import silhouette_score # başarı oranımızı 
```

```python
df.columns
```

```python
for col in df.select_dtypes("object").columns[1:]:
    df[col]  = df[col].astype("str")
    df[col]  = df.select_dtypes("object")[col].str.replace(',','')
    df[col]  = df[col].astype("float")
```

```python
x= df.drop(["Country"], axis=1)
```

```python
x.head()
```

```python
model=KMeans(n_clusters=4)
```

```python
model=model.fit(x)
```

```python
tahmin=model.predict(x)
```

```python
tahmin
```

```python
x["cluster"]=tahmin
```

```python
x.head()
```

```python
x.shape
```

```python
silhouette_score(x,tahmin)
```

```python
allscore=[]
allclusters=[]
kırılma=[]
def clust(kactane):
    
    for i in range(1,kactane+1):
        i=i+1
        model=KMeans(n_clusters=i)
        model=model.fit(x)
        tahmin=model.predict(x)
        score=silhouette_score(x,tahmin)
        uzaklik=model.inertia_
        print("Cluster ",i," score ",score)
        allscore.append(score)
        allclusters.append(i)
        kırılma.append(uzaklik)
clust(10)
```

```python
plt.xlabel("k")
plt.ylabel("kırılma")
plt.plot(allclusters,kırılma)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df["Aircraft Strength value"],df["Attack Aircraft Strength value"],c=tahmin)
plt.xlabel("Uçak Gücü değeri")
plt.ylabel("Saldırı Uçağı Gücü değeri")
plt.colorbar()
plt.grid()
```
